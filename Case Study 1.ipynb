# Step -9 Outlier Detection

```python
sns.boxplot(y=df1['AveragePrice'])
plt.show()
```

```python
# so we need to handle it(between 25 tile and 75 tile)
```

## For Average Price Column

```python
q1 = df1['AveragePrice'].quantile(.25)
q1
```

```python
q3 = df1['AveragePrice'].quantile(.75)
q3
```

```python
iqr = q3 - q1
iqr
```

```python
Lower_Limit = q1 - 1.5 * iqr
Lower_Limit
```

```python
Upper_Limit = q3 + 1.5 * iqr
Upper_Limit
```

```python
df[df1['AveragePrice'] < Lower_Limit]
```

```python
df[df1['AveragePrice'] > Upper_Limit]
```

```python
so we need to handle the outer treshold values
```

```python
df1['AveragePrice'] = df1['AveragePrice'].clip(lower=Lower_Limit, upper=Upper_Limit)
df1['AveragePrice']
```

```python
df[df1['AveragePrice'] > Upper_Limit]
```

```python
sns.boxplot(y=df1['AveragePrice'])
plt.show()
```

```python
q1 = df1[['Total Volume','Small_Volume','Large_Volume','xLarge_Volume',
                     'Total Bags','Small Bags','Large Bags','XLarge Bags']].quantile(.25)
q1
```

```python
q3 = df1[['Total Volume','Small_Volume','Large_Volume','xLarge_Volume',
                     'Total Bags','Small Bags','Large Bags','XLarge Bags']].quantile(.75)
q3
```

```python
iqr = q3 - q1
iqr
```

```python
Lower_Limit = q1 - 1.5 * iqr
Lower_Limit
```

```python
Upper_Limit = q3 + 1.5 * iqr
Upper_Limit
```

```python
df[df1[['Total Volume','Small_Volume','Large_Volume','xLarge_Volume',
                     'Total Bags','Small Bags','Large Bags','XLarge Bags']] > Upper_Limit]
```

```python
df[df1[['Total Volume','Small_Volume','Large_Volume','xLarge_Volume',
                     'Total Bags','Small Bags','Large Bags','XLarge Bags']] < Lower_Limit]
```

```python
df1[['Total Volume','Small_Volume','Large_Volume','xLarge_Volume',
     'Total Bags','Small Bags','Large Bags','XLarge Bags']] = df1[['Total Volume','Small_Volume','Large_Volume','xLarge_Volume',
     'Total Bags','Small Bags','Large Bags','XLarge Bags']].clip(lower=Lower_Limit, upper=Upper_Limit, axis = 1)
```
